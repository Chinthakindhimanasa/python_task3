```python
# Q4: grade_with_logical_safe.py

user_input = input("Enter your marks: ").strip()  # Remove spaces

if user_input == "":
    print("You did not enter any marks!")
else:
    try:
        marks = float(user_input)
        
        # Using logical operators (and / or)
        if marks >= 70 and marks < 80:
            print("Grade: B")
        elif marks >= 60 and marks < 70:
            print("Grade: C")
        elif marks >= 50 or marks == 50:  # 'or' used for demonstration
            print("Grade: D")
        else:
            print("Grade: F")
    
    except ValueError:
        print("Invalid input! Please enter numeric marks.")
```

```python
# grade_calculator.py

def calculate_grade(marks):
    """
    Determines the grade and message based on marks.
    Includes nested conditions and logical operators.
    """
    if marks < 0 or marks > 100:  # Requirement 5: handle invalid marks
        return "Invalid marks! Enter a number between 0 and 100."

    # Requirement 7: nested conditions for business rules
    if marks >= 90:
        return "A+ Grade - Excellent!"
    elif marks >= 80:
        return "A Grade - Very Good!"
    elif marks >= 70:
        if marks >= 75:
            return "B+ Grade - Almost A"
        else:
            return "B Grade"
    elif marks >= 60:
        if marks >= 65:
            return "C+ Grade - Almost B"
        else:
            return "C Grade"
    elif marks >= 50:
        # Requirement 4: using logical operator OR
        if marks == 50 or marks == 51:
            return "D Grade - Just passed"
        else:
            return "D Grade"
    else:
        return "F Grade - Failed"

# Requirement 9: allow testing multiple inputs
try:
    n = int(input("How many marks do you want to enter? "))
except ValueError:
    print("Invalid input! Please enter a numeric value.")
    n = 0

for i in range(n):
    try:
        marks = float(input(f"Enter marks for input {i+1}: "))
        print(calculate_grade(marks))
    except ValueError:
        print("Invalid input! Please enter numeric marks.")

```
